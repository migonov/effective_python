## Zadanie 2

Zadanie należy wykonać wykorzystując program `BagOfWords` z poprzednich zajęć.


Przerób ją tak, żeby interpunkcja, cyfry i wszelkie inne znaki nie przeszkadzały w parsowaniu tekstu. Uruchom ja na [tekście hamleta](http://www.gutenberg.org/cache/epub/1787/pg1787.txt). Ile razy występuje słowo hamlet? Jak brzmi 10 najczęściej występujących słów?

```python
from collections import defaultdict
import string
import re

class BagOfWords(object):
    def __init__(self, value=None):
        self.dict = defaultdict(self.zero)
        if isinstance(value, str):
            for word in value.split():
                word = word.translate(word.maketrans("", "", string.punctuation))
                word = re.sub('[^a-zA-Z\s]', '', word)
                self.dict[word.lower()] += 1
        elif value is not None:
            for line in value:
                line = line.translate(line.maketrans("", "", string.punctuation))
                line = re.sub('[^a-zA-Z\s]', '', line)
                line = re.sub('\s+', ' ', line)
                for word in line.split():
                    self.dict[word.lower()] += 1
            value.close()
            
    def zero(self):
        return 0
    
    def __getitem__(self, key):
        return self.dict[key.lower()]

    def __setitem__(self, key, value):
        self.dict[key.lower()] = value

    def __iter__(self):
        words = defaultdict(lambda: set())
        for k, v in self.dict.items():
            words[v].add(k)
        for count in reversed(sorted(words.keys())):
            for word in words[count]:
                yield word
    
    def __add__(self, other):
        new_bag = BagOfWords()
        for k in other:
            new_bag[k] += other[k]
        for k, v in self.dict.items():
            new_bag[k] += v
        return new_bag
    
    def __repr__(self):
        return '\n'.join([f'{key}:{val}' for key, val in self.dict.items()])
    
hamlet_bag = BagOfWords(open("hamlet.txt"))
print(f'***Liczba wystąpień Hamlet: {hamlet_bag["Hamlet"]}***')
print("\n***Top 10 słów***")
for index, word in enumerate(hamlet_bag):
    if index >= 10:
        break
    print(f'{word}: {hamlet_bag[word]}')
print("\n***Całość***")
print(hamlet_bag)
```

## Zadanie 4

Wykorzystując [https://gist.github.com/pamelafox/986163](https://gist.github.com/pamelafox/986163) podaj aktualną godzinę we:
- wszystkich krajach, wyświetlając je zgrupowane względem kontynentów,
- przeprowadź symulacyjne wyświetlanie kolejnych krajów w miarę jak w danym kraju wybija północ, opóżnienie wyświetlania ustaw proporcjonalne do realnego czasu

```python
from datetime import datetime
from collections import defaultdict
from pytz import timezone
from time import sleep
import countryinfo
  
m = defaultdict(lambda: list())
for continent_timezones in sorted(countryinfo.countries, key = lambda x : x['continent']):
    print(continent_timezones['continent'], continent_timezones['name'])
    for j in continent_timezones['timezones']:
        curr_time = datetime.now(timezone(j)).strftime("%H:%M:%S")
        print(j, curr_time)
        m[curr_time].append(j)
    print()
    
for time, countries in sorted(m.items()):
    print(time)
    for x in countries:
        print(x)
    print()
    sleep(1)
```

## Zadanie 5

Dla klasy `BagOfWords` napisz metody `save` oraz `load` wykorzystujące `json`'a do zapisu i odczytu danych.

```python
from collections import defaultdict
import string
import re
import json

class BagOfWords(object):
    def __init__(self, value=None):
        self.dict = defaultdict(lambda: 0)
        if isinstance(value, str):
            for word in value.split():
                word = word.translate(word.maketrans("", "", string.punctuation))
                word = re.sub('[^a-zA-Z\s]', '', word)
                self.dict[word.lower()] += 1
        elif value is not None:
            for line in value:
                line = line.translate(line.maketrans("", "", string.punctuation))
                line = re.sub('[^a-zA-Z\s]', '', line)
                line = re.sub('\s+', ' ', line)
                for word in line.split():
                    self.dict[word.lower()] += 1
            value.close()
    
    def save(self, filename):
        with open(filename, 'w') as file:
            json.dump(self.dict, file)
    
    def load(self, filename):
        with open(filename, 'r') as file:
            self.dict = json.load(file)
    
    def __getitem__(self, key):
        return self.dict[key.lower()]

    def __setitem__(self, key, value):
        self.dict[key.lower()] = value

    def __iter__(self):
        words = defaultdict(lambda: set())
        for k, v in self.dict.items():
            words[v].add(k)
        for count in reversed(sorted(words.keys())):
            for word in words[count]:
                yield word
    
    def __add__(self, other):
        new_bag = BagOfWords()
        for k in other:
            new_bag[k] += other[k]
        for k, v in self.dict.items():
            new_bag[k] += v
        return new_bag
    
    def __repr__(self):
        return '\n'.join([f'{key}:{val}' for key, val in self.dict.items()])
    
filename = "test.json"
bag = BagOfWords("ala ma kota")
bag.save(filename)
bag2 = BagOfWords()
bag2.load(filename)

print(bag)
print("---------------------")
print(bag2)
```
